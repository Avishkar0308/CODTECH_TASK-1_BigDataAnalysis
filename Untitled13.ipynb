```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder \
    .appName("CODTECH_BigDataAnalysis_MallCustomers") \
    .getOrCreate()

spark
```

```python
# Load the uploaded dataset
df = spark.read.csv("/content/Mall_Customers.csv", header=True, inferSchema=True)

# Show first few rows
df.show(5)

# Display schema
df.printSchema()
```

```python
# Count total records
print("Total Rows:", df.count())

# Summary statistics
df.describe().show()

# Check distinct genders
df.select("Gender").distinct().show()
```

```python
from pyspark.sql.functions import col, avg, min, max, count

# Average Spending Score and Income by Gender
df.groupBy("Gender") \
  .agg(avg("Spending Score (1-100)").alias("Avg_SpendingScore"),
       avg("Annual Income (k$)").alias("Avg_Income")) \
  .show()

# Top 5 High Income Customers
df.orderBy(col("Annual Income (k$)").desc()).show(5)
```

```python
# Create a new column "Income Level"
df = df.withColumn(
    "Income_Level",
    col("Annual Income (k$)") > 70
)

# Show few rows
df.select("CustomerID", "Annual Income (k$)", "Income_Level").show(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert small sample to Pandas for plotting
sample = df.limit(200).toPandas()

plt.figure(figsize=(8,5))
plt.scatter(sample["Annual Income (k$)"], sample["Spending Score (1-100)"], c='orange')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Income vs Spending Score (Mall Customers)")
plt.show()
```
